### Highest growth rate for Investing

```python
all_companies = data.columns[1:]

growth_on_all_stocks = data[all_companies].pct_change() * 100

average_growth_on_all_stocks = growth_on_all_stocks.mean()

average_growth_on_all_stocks.sort_values(ascending = False).head(10)
```

### Highest risk companies for Investing

```python
volatility_all_stocks = data[all_companies].std()

volatility_all_stocks.sort_values(ascending = False).head(10)
```

### Highest return on Investments

```python
all_initial_stock_prices = data[all_companies].iloc[0]
all_final_stock_prices = data[all_companies].iloc[-1]

roi_on_all_stocks = ((all_final_stock_prices - all_initial_stock_prices) / all_initial_stock_prices) * 100

roi_on_all_stocks.sort_values(ascending = False).head(10)
```

## Creating a Mutual Fund Plan Based on High ROI and Low Risk

To design an effective strategy for selecting companies with high returns and controlled risk, we combine Return on Investment (ROI) and volatility (standard deviation) as key evaluation metrics. The objective is to identify stocks that deliver strong long-term returns while maintaining relatively stable price movements.

The mutual fund creation process follows these steps:

**1. Define ROI and Volatility Thresholds**

Establish benchmark values for ROI and volatility to filter companies that demonstrate above-average returns with below-average risk.

**2. Rank Companies by Performance and Risk**

Rank all companies based on their ROI and volatility metrics to evaluate their growth potential and stability.

**3. Assign Investment Weights**

Allocate higher investment weights to companies with lower volatility, ensuring risk is balanced across the portfolio while preserving return potential.

We begin by defining ROI and volatility thresholds and selecting companies that satisfy the criteria of high return and low risk, forming the foundation of the mutual fund portfolio.

```python
roi_threshold = roi_on_all_stocks.median()
volatility_threshold = volatility_all_stocks.median()

selected_companies = roi_on_all_stocks[
    (roi_on_all_stocks > roi_threshold) &
    (volatility_all_stocks < volatility_threshold)
]

selected_companies.sort_values(ascending = False)
```

To balance investments, we use inverse volatility weighting, where companies with lower volatility receive higher weights. This helps control risk while allocating investments across selected companies.

```python
selected_vol = volatility_all_stocks[selected_companies.index]

inverse_vol = 1 / selected_vol

investment_ratios = inverse_vol/inverse_vol.sum()

investment_percentages = investment_ratios * 100

(investment_percentages.sort_values(ascending=False))
```

# Calculating Expected Returns

Now, letâ€™s calculate the expected returns a person will get from our mutual fund if he/she invests â‚¹5000 every month.

To calculate the expected value a person will accumulate over 1 year, 3 years, 5 years, and 10 years through the mutual fund plan, we can follow these steps:

1. Assume the person is investing 5000 rupees every month & increasing the investment amount by 10% every year.
2. Use the expected ROI from the mutual fund companies to simulate the growth over time.
3. Compute the compounded value of the investments for each period (1y, 3y, 5y, and 10y).
4. Visualize the accumulated value over these periods.

```python
def sip_future_value(monthly_investment, monthly_return, years, step_up=0.10):
  total_value = 0
  current_monthly = monthly_investment
  for year_count in range(1, years + 1):
    for month in range(12):
      total_value = (total_value + current_monthly) * (1 + monthly_return)
    current_monthly *= (1 + step_up)
  return total_value


selected_stocks = selected_companies.head(5)
expected_annual_return = roi_on_all_stocks[selected_companies.index].mean() / 100
monthly_return = expected_annual_return / 12

print(expected_annual_return)
print(monthly_return)

monthly_investment = 5000
years_list = [1,3,5,10]

sip_results = {}

for year in years_list:
  sip_results[f"{year} Years"] = sip_future_value( monthly_investment, monthly_return, year)

print(sip_results)
```

```python

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=[f"{year} year{'s' if year > 1 else ''}" for year in years_list],
    y=[sip_results[f"{year} Years"] for year in years_list],
    mode='lines+markers',
    line=dict(color='blue'),
    marker=dict(size=8),
    name='Future Value'
))

fig.update_layout(
    title="Expected Value of Monthly SIP with 10% Annual Increase (â‚¹5000 Starting)",
    xaxis_title="Investment Period",
    yaxis_title="Future Value (â‚¹)",
    template="plotly_white",
    hovermode="x unified"
)

fig.show()
```

After 1 year, the accumulated value is around â‚¹71,000 and by 5 years, it grows to over â‚¹821,100. The long-term benefit is evident, with the investment growing to nearly â‚¹5,054,931 over 10 years, which emphasises the value of consistent investing and compounding over time for long-term investors.
